3.Find the most frequent value and their indice(s) in the following arrays: 
i.
x = np.array([1,2,3,4,5,1,2,1,1,1]) 
ii.
4.
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ]) 

```python

x = np.array([1,2,3,4,5,1,2,1,1,1])
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])

def function(arr, name):
    vals, counts = np.unique(arr, return_counts=True)
    max_count = counts.max()

    modes = vals[counts == max_count]

    print(f"Results for array {name}:")
    for mode in modes:
        indices = np.where(arr == mode)[0]
        print(f"Value: {mode}, Indices: {indices}")
    print("")

function(x, "x")
function(y, "y")
```


4. For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), 
find 

i. Sum of all elements 

ii. Sum of all element’s row-wise 

iii.Sum of all element’s column-wise 


```python

gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

# i. Sum of all elements
print(f" Sum of all elements: {gfg.sum()}")

# ii. Sum of all Elements row-wise
print(f" Row Wise Sum: {gfg.sum(axis=1)}")

# iii. Sum of all Elements column-wise
print(f" Column-wise Sum: {gfg.sum(axis=0)}")
```
